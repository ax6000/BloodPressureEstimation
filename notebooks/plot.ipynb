```python
import matplotlib.pyplot as plt

# 各区間と設定
ranges = [(50, 100), (100, 120), (120, 140), (140, 180)]
labels = ['0', '1', '2', '3']
colors = ['skyblue', 'lightgreen', 'gold', 'salmon']
sample_counts = [137448,268709,202811,112472]  # ここに各範囲のサンプル数を指定

fig, ax = plt.subplots(figsize=(10, 2.5))

# 横棒とラベル＋サンプル数を描画
for i, ((start, end), label, count, color) in enumerate(zip(ranges, labels, sample_counts, colors)):
    ax.barh(0, width=end - start, left=start, height=0.6, color=color, edgecolor='black')
    x_center = (start + end) / 2
    ax.text(x_center, 0.1, f'Label {label}', ha='center', va='bottom', fontsize=12, fontweight='bold')
    ax.text(x_center, -0.15, f'n = {count}', ha='center', va='top', fontsize=12)

# 軸とスタイルの調整
ax.set_xlim(50, 180)
ax.set_ylim(-0.5, 0.5)
ax.set_yticks([])
ax.set_xlabel("SBP (mmHg)", fontsize=12)
ax.tick_params(axis='x', labelsize=12)
# ax.set_title("SBP Classification Ranges with Sample Counts", fontsize=14)

plt.tight_layout()
plt.show()
```
